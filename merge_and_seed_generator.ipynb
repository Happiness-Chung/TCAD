```python
import pandas as pd
```

```python
train = pd.read_csv("../ICASC-/data/CheXpert-v1.0-small/selected_train.csv")
test = pd.read_csv("../ICASC-/data/CheXpert-v1.0-small/selected_test.csv")
```

```python
df = pd.concat([train, test], axis=0)
```

```python
from sklearn.model_selection import train_test_split
seed = 3
X_train, X_test = train_test_split(df['Path'], test_size=0.2, random_state = seed)
sublist = ['train_subjects']+list(X_train)+['test_subjects']+list(X_test)
with open(f"./splits/CheXpert/CheXpert_split{seed}.txt", mode="w") as file:
    file.write('\n'.join(sublist))
```
