```python
seed = 2

X_train, X_test = train_test_split(df['Path'], test_size=0.15, random_state = seed)
X_train, X_valid = train_test_split(X_train, test_size=0.175, random_state = seed)
print(len(X_train), len(X_valid), len(X_test))

sublist = ['train_subjects']+list(X_train)+['valid_subjects']+list(X_valid)+['test_subjects']+list(X_test)
with open(f"./splits/CheXpert/CheXpert_split{seed}.txt", mode="w") as file:
    file.write('\n'.join(sublist))
```
